```python
import requests
import json
```

```python
repsonse = requests.get('https://api.stackexchange.com/2.3/questions?order=desc&sort=activity&site=stackoverflow')

print(repsonse,'\n')
print(repsonse.json(),'\n')
print(repsonse.json()['items'])
```

```python
for data in repsonse.json()['items']: #this is a list of dictionary
    print(data['title']) #print the component 'title' of each iterated dictionary
    print(data['link'])
```
